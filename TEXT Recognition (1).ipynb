# <center> TEXT DETECTION <center>

Text detection is the process of detecting the text present in the image, followed by 
surrounding it with a rectangular bounding box. Text detection can be carried out using 
image-based techniques or machine learning-based techniques. In image-based 
techniques, an image is segmented into multiple segments, and statistical features of connected 
components are utilised to form the text. Whereas, Machine learning approaches use support 
vector machines and convolutional neural networks to classify the components into text and 
non-text.

Implement a text detection and extraction model using OpenCV and OCR. The necessary 
steps that you need to perform are: 
1. Image preprocessing 
2. Find possible contours that can represent the textual areas. 
3. Apply optical character recognition (using python-tesseract, google OCR engine.

```python
image_path=r'C:\Users\HP\Downloads\text22.jpeg'
img_cv = cv2.imread(r'C:\Users\HP\Downloads\text22.jpeg')

# By default OpenCV stores images in BGR format and since pytesseract assumes RGB format,
# we need to convert from BGR to RGB format/mode:
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
print(pytesseract.image_to_string(img_rgb))
# OR
#img_rgb = Image.frombytes('RGB', img_cv.shape[:2], img_cv, 'raw', 'BGR', 0, 0)
#print(pytesseract.image_to_string(img_rgb))
# Step 1: Image Preprocessing
def preprocess_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to create a binary image
    _, binary_image = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Apply noise reduction
    denoised_image = cv2.fastNlMeansDenoising(binary_image, None, h=10, hForColor=15, templateWindowSize=7, searchWindowSize=21)

    return denoised_image

# Step 2: Find Textual Contours
def find_text_contours(image):
    # Find contours in the image
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    textual_contours = []

    for contour in contours:
        # Filter contours based on area or other criteria if needed
        if cv2.contourArea(contour) > 100:
            textual_contours.append(contour)

    return textual_contours

# Step 3: Apply OCR using pytesseract
def extract_text_from_contours(image, contours):
    extracted_text = []

    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        roi = image[y:y+h, x:x+w]

        # Use pytesseract to perform OCR on the region of interest (ROI)
        text = pytesseract.image_to_string(roi, config='--psm 6')  # psm 6 assumes a single block of text

        if text.strip():
            extracted_text.append((x, y, text))

    return extracted_text
```
